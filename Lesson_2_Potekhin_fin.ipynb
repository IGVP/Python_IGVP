```python
# Импортируем библиотеку Pandas
import pandas as pd
```

```python
# Загружаем данные 'marketing_campaign.csv' в Python c помощью бибилиотеки Pandas
df_market_camp = pd.read_csv('marketing_campaign.csv')
# Посмотрим количество строк и столбцов в датафрейме
print(f'Размер датафрейма: {df_market_camp.shape}')
# Возвратим первые 3 строки загруженного датафрейма
df_market_camp.head(3)
```

```python
# Загружаем данные 'users.csv' в Python c помощью бибилиотеки Pandas
df_users = pd.read_csv('users.csv')
# Посмотрим празмер датафрейма
print(f'Размер датафрейма: {df_users.shape}')
# Возвратим первые 3 строки загруженного датафрейма
df_users.head(3)
```

```python
# Загружаем данные 'users.csv' в Python c помощью бибилиотеки Pandas
df_subscribe = pd.read_csv('subscribers.csv')
# Посмотрим количество строк и столбцов в датафрейме
print(f'Размер датафрейма: {df_subscribe.shape}')
# Возвратим первые 3 строки загруженного датафрейма
df_subscribe.head(3)
```

```python
# Объеденим датафреймы
df = df_market_camp.merge(df_users, on = 'user_id', how = 'left')\
    .merge(df_subscribe, on = 'user_id', how = 'left')
# Смотрим размер датасета
print(f'Размер датасета: {df.shape}')
# Выводим датасет (первые три строки)
df.head(3)
```

```python
# Общая статистика датасета df_market_research
df.info()
```

```python
# Описательная статистика для нечисловых столбцов датасета
df.describe(include=['object', 'bool'])
```

Определить эффективность маркетинговых каналов по привлечению платящих игроков.

Проверка эффективности канала по удержанным пользователям.

```python
# С какого канала пришел пользователь (эффективность маркетинговых каналов)
df['subscribing_channel'].value_counts()
```

```python
# Проверим эффективность маркетингового канала в отношении удержанных пользоватлей
df.loc[df['is_retained'] == 1, 'subscribing_channel'].value_counts()
```

Смамым эффективным каналом является Instagram

Определить самую раннюю дату подписки на сервис.

```python
# Перевод строки в дату
df['date_subscribed'] = pd.to_datetime(df['date_subscribed'])
# Выбор минимальной даты и ее отображение
print('Дата перводй подписки: ', min(df['date_subscribed']))
```

Самая ранняя дата подписки - 01.01.2018

Определить портрет аудитории удержанных подписчиков (по возрасту и языку).

```python
# Формируем датасет удержанных пользователей, включаем возрастную группу и язык 
df_users_retained = df.loc[df['is_retained'] == 1, ['language_preferred','age_group']]
# Выводим размер таблицы
print(f'Размер таблицы: {df_users_retained.shape}\n')
# Смотрим портрет пользователей
print(f'Просмотр портета через value_counts\n\n{df_users_retained.value_counts()}')
print('\nПортрет аудитории удержанных подписчиков (по возрасту и языку)\n')
pd.crosstab(index = df_users_retained['language_preferred'], columns = df_users_retained['age_group'], margins = True)
```
